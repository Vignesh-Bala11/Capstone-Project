```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from config import *
import psycopg2
```

```python
conn = psycopg2.connect(database=DATABASE, user=USER, password=PASSWORD, host=HOST, port= PORT)
cursor = conn.cursor()
```

```python
# Loading in Results dataset as Dataframe
cursor.execute("Select * FROM results LIMIT 0")
colnames = [desc[0] for desc in cursor.description]
cursor.execute("SELECT * FROM results")
data = cursor.fetchall()
results = pd.DataFrame(data, columns=colnames)
results.head()
```

```python
# Loading in Races dataset as Dataframe
cursor.execute("Select * FROM races LIMIT 0")
colnames = [desc[0] for desc in cursor.description]
cursor.execute("SELECT * FROM races")
data = cursor.fetchall()
races = pd.DataFrame(data, columns=colnames)
races.head()
```

```python
# Loading in Driver  dataset as Dataframe
cursor.execute("Select * FROM drivers LIMIT 0")
colnames = [desc[0] for desc in cursor.description]
cursor.execute("SELECT * FROM drivers")
data = cursor.fetchall()
drivers = pd.DataFrame(data, columns=colnames)
drivers.head()
```

```python
# Loading In Constructor Standing dataset as DataFrame
cursor.execute("Select * FROM constructor_standings LIMIT 0")
colnames = [desc[0] for desc in cursor.description]
cursor.execute("SELECT * FROM constructor_standings")
data = cursor.fetchall()
constructor_standings = pd.DataFrame(data, columns=colnames)
constructor_standings.head()
```

```python
# Loading In Constructors dataset as DataFrame
cursor.execute("Select * FROM constructors LIMIT 0")
colnames = [desc[0] for desc in cursor.description]
cursor.execute("SELECT * FROM constructors")
data = cursor.fetchall()
constructors = pd.DataFrame(data, columns=colnames)
constructors.head()
```

```python
# Loading In Qualifying dataset  as DataFrame
cursor.execute("Select * FROM qualifying LIMIT 0")
colnames = [desc[0] for desc in cursor.description]
cursor.execute("SELECT * FROM qualifying")
data = cursor.fetchall()
quals = pd.DataFrame(data, columns=colnames)
quals.head()
```

```python
# Loading in Driver Standings dataset as DataFrame
cursor.execute("Select * FROM driver_standings LIMIT 0")
colnames = [desc[0] for desc in cursor.description]
cursor.execute("SELECT * FROM driver_standings")
data = cursor.fetchall()
driver_standings = pd.DataFrame(data, columns=colnames)
driver_standings.head()
```

```python
# Loading in Circuits Standings dataset as DataFrame
cursor.execute("Select * FROM circuits LIMIT 0")
colnames = [desc[0] for desc in cursor.description]
cursor.execute("SELECT * FROM circuits")
data = cursor.fetchall()
circuits = pd.DataFrame(data, columns=colnames)
circuits.head()
```

```python
df1 = pd.merge(races,results,how='inner',on=['raceId'])
df2 = pd.merge(df1,quals,how='inner',on=['raceId','driverId','constructorId'])
df3 = pd.merge(df2,drivers,how='inner',on=['driverId'])
df4 = pd.merge(df3,constructors,how='inner',on=['constructorId'])
df5 = pd.merge(df4,circuits,how='inner',on=['circuitId'])
df5.head()
```
